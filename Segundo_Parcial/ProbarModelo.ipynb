```python
# Prediccion
from ultralytics import YOLO

# Load a model
# model = YOLO('yolov8x-seg.pt')  # load an official model
model = YOLO('./runs_30epochs/segment/train/weights/best.pt')  # load a custom model

# Predict with the model
results = model('./123.jpg')  # predict on an image
print(results)
```

```python
from PIL import Image
from ultralytics import YOLO
from PIL import ImageDraw

# Load a pretrained YOLOv8n model
model = YOLO('./runs_30epochs/segment/train/weights/best.pt')

# Open an image using PIL
source = Image.open('./123.jpg')

# Run inference on the source
results = model(source)  # list of Results objects

# Draw bounding boxes on the image
draw = ImageDraw.Draw(source)
for result in results.xyxy[0]:
    # 'xyxy' contains the bounding box coordinates and class confidence
    # Extract the coordinates
    x_min, y_min, x_max, y_max = map(int, result[:4])
    
    # Draw a rectangle on the image
    draw.rectangle([x_min, y_min, x_max, y_max], outline="red", width=3)

# Show the image with bounding boxes
source.show()
```
